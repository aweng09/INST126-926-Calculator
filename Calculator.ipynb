```python
# Boolean variables initialized

ask_need = True
ask_cont = True
ask_save = False
valid_op = True

# Begin calculator program, as long as the user wants to continue using it

while ask_cont is True:
    
    # Asks which operation the user wants to perform, and allows the user to continue if the selection is valid
    
    ask_op = input("What operation would you like to perform? Choose a number. 1. Addition 2. Subtraction 3. Multiplication 4. Division\n")
    
    if (ask_op != "1") and (ask_op != "2") and (ask_op != "3") and (ask_op != "4"):
        valid_op = False
        while valid_op is False:
            print("Sorry! I didn't understand that, please try again.")
            ask_op = input("What operation would you like to perform? Choose a number. 1. Addition 2. Subtraction 3. Multiplication 4. Division\n")
            if (ask_op == "1") or (ask_op == "2") or (ask_op == "3") or (ask_op == "4"):
                    valid_op = True
        
    # Asks for input of the first number if the user does not save a previous answer, tells you off if you didn't enter an integer
    
    if ask_save is False:
        
        num = int(input("Please enter the first number, bearing in mind order. (For instance, if it is division, the dividend first. With subtraction, the number subtracted from first.)\n")) 

    # Asks for the next number in the operation, as long as the user still needs to enter numbers
    
    while ask_need is True:
            
        temp = int(input("Now, the next number!\n"))
        
        if ask_op == "1":
            num = num + temp
        elif ask_op == "2":
            num = num - temp
        elif ask_op == "3":
            num = num * temp
        elif ask_op == "4":
            if temp == 0:
                print("Okay, very funny; you can't split " + str(num) + " cookies with 0 friends. Just ask Siri.")
            else:
                num = num / temp
        
        ask_need_str = input("Need another number? Yes or No.\n")
        if ask_need_str == "No":
            ask_need = False
                        
    # Asks the user if they would like to save their answer to use in the next operation
    
    ask_save_str = input("Your answer is: " + str(num) + ". Save answer? Yes or No.\n")
    
    if ask_save_str == "No":
        ask_save = False
    else:
        ask_save = True
        
    # Asks user if they would like to continue to use the program    
    
    ask_cont_str = input("Would you like to perform another operation? Yes or No.\n")
    if ask_cont_str == "No":
        ask_cont = False
    else:
        ask_cont = True
        ask_need = True
        
print("Calculator exited.")
```
